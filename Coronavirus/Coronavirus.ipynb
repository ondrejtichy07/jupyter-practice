```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
path = 'data.csv'
```

```python
data = pd.read_csv(path, sep=';', header=None, usecols=[0,1,2,3], names=['Date','Country','NewCases','NewDeaths'])
data_set = data.drop(data.index[0])
data_set['NewCases'] = pd.to_numeric(data_set['NewCases'])
data_set['NewDeaths'] = pd.to_numeric(data_set['NewDeaths'])
czechia = data_set[data_set.Country == 'Czech Republic'].head(15).sort_values(by='Date', ascending=1)
```

```python
print(czechia)
```

```python
data_italy = pd.read_csv(path, sep=';', header=None, usecols=[0,1,2,3], names=['Date','Country','NewCases','NewDeaths'])
data_set_italy = data_italy.drop(data.index[0])
data_set_italy['NewCases'] = pd.to_numeric(data_set_italy['NewCases'])
data_set_italy['NewDeaths'] = pd.to_numeric(data_set_italy['NewDeaths'])
italy = data_set_italy[data_set_italy.Country == 'Italy'].head(15).sort_values(by='Date', ascending=1)
```

```python
plt.plot(italy['Date'], italy['NewCases'], color='blue')
plt.plot(italy['Date'], italy['NewDeaths'], color='red')

plt.title('Coronavirus cases in Italy per day', fontsize=13)
plt.grid(linestyle='-', linewidth=.4)

plt.xticks(rotation=65, fontsize=10)

plt.xlabel('Date', fontsize=10)
plt.ylabel('Cases per day', fontsize=10)

plt.show()
```

```python
plt.plot(czechia['Date'], czechia['NewCases'], color='blue')
plt.plot(czechia['Date'], czechia['NewDeaths'], color='red')

plt.title('Coronavirus cases in the Czech Republic per day', fontsize=13)
plt.grid(linestyle='-', linewidth=.4)

plt.xticks(rotation=65, fontsize=10)

plt.xlabel('Date', fontsize=10)
plt.ylabel('Cases per day', fontsize=10)

plt.show()
```
