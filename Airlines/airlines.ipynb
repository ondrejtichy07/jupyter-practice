```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('data.csv')
data.sort_values(by=['fatal_accidents_85_99'], ascending=False)
```

```python
df0=pd.DataFrame(data, columns=['airline','fatal_accidents_00_14','fatalities_00_14'])
df=df0.sort_values(by=['fatalities_00_14'], ascending=False)
print(df)
```

```python
x1 = data['airline']
y1 = data['fatalities_00_14']
my_color = np.where(y1>0,'green', 'red')

plt.figure(figsize=(25,4))
plt.scatter(x1, y1, color=my_color)
plt.xticks(rotation=90, fontsize=15)
plt.grid(linestyle='-', linewidth=.4)
plt.title('Airline fatalities between 2000 and 2014')

plt.show()
```
